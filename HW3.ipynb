```python
#Bai 7 Toi uu
a = int(input("Nhập số nguyên dương a: "))

count = 0
for i in range(1, int(a**0.5) + 1):
    if a % i == 0:
        if i == a // i:
            count += 1
        else:
            count += 2

print("Số ước dương của", a, "là:", count)
```
