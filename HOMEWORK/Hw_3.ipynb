```python
# Bài 9
a = int(input())
is_perfect_square = False
i = 1
while i * i <= a:
    if i * i == a:
        is_perfect_square = True
        break
    i += 1
    
if is_perfect_square:
    print("La so chinh phuong")
else:
    print("Khong phai so chinh phuong")
```

```python
# Bài 10
a = int(input())
total_divisors = 0
for i in range(1, a):
    if a % i == 0:
        total_divisors += i

if total_divisors == a:
    print("La so hoan hao")
else:
    print("Khong phai so hoan hao")
```

```python
# Bài 14
val = int(input())
if val == -1:
    print("Khong co so nao duoc nhap")
else:
    max_val = val
    min_val = val
    while True:
        val = int(input())
        if val == -1:
            break
        if val > max_val:
            max_val = val
        if val < min_val:
            min_val = val
    print(max_val, min_val)
```

```python
# Bài 15
n = int(input())
count = 0
temp = n
if temp == 0:
    count = 1
else:
    while temp > 0:
        temp //= 10
        count += 1
print(count)
```

```python
# Bài 16
n = int(input())
even_count = 0
odd_count = 0
while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
    n //= 10
print(even_count, odd_count)
```

```python
# Bài 17
n = int(input())
total = 0
while n > 0:
    total += n % 10
    n //= 10
print(total)
```

```python
# Bài 18
n = int(input())
if n < 1:
    print("Khong phai dang 3^k")
else:
    temp = n
    while temp > 1:
        if temp % 3 != 0:
            break
        temp //= 3
    
    if temp == 1:
        print("La dang 3^k")
    else:
        print("Khong phai dang 3^k")
```

```python
# Bài 19
A = int(input())
f1, f2 = 1, 1
res = 1
if A < 1:
    print(0)
else:
    while True:
        curr = f1 + f2
        if curr > A:
            res = f2
            break
        f1 = f2
        f2 = curr
        res = f2
    print(res)
```

```python
# Bài 20
n = int(input())
count = 0
while n > 0:
    digit = n % 10
    if digit == 2 or digit == 3 or digit == 5 or digit == 7:
        count += 1
    n //= 10
print(count)
```
