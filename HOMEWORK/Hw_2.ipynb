```python
# Bài 3
import math
a = float(input())
b = float(input())
c = float(input())

if a + b > c and a + c > b and b + c > a:
    type_tri = "Tam giac thuong"
    if a == b == c:
        type_tri = "Tam giac deu"
    elif a == b or b == c or a == c:
        if a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
             type_tri = "Tam giac vuong can"
        else:
             type_tri = "Tam giac can"
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        type_tri = "Tam giac vuong"
    
    p = (a + b + c) / 2
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
    chu_vi = a + b + c
    
    print(type_tri)
    print(dien_tich)
    print(chu_vi)
else:
    print("Khong la ba canh cua mot tam giac")
```

```python
# Bài 5
import math
a = float(input())
b = float(input())
c = float(input())

if a == 0:
    if b == 0:
        if c == 0:
            print("Vo so nghiem")
        else:
            print("Vo nghiem")
    else:
        print(-c / b)
else:
    delta = b*b - 4*a*c
    if delta > 0:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print(x1, x2)
    elif delta == 0:
        print(-b / (2*a))
    else:
        print("Vo nghiem")
```

```python
# Bài 8
a = float(input())
b = float(input())
m = float(input())
c = float(input())
d = float(input())
n = float(input())

D = a*d - c*b
Dx = m*d - n*b
Dy = a*n - c*m

if D != 0:
    x = Dx / D
    y = Dy / D
    print(x, y)
else:
    if Dx != 0 or Dy != 0:
        print("Vo nghiem")
    else:
        print("Vo so nghiem")
```
