```python
# Bài 2
import math

a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))

# Giả sử đường kính hình tròn bằng chiều rộng (cạnh nhỏ hơn)
r = min(a, b) / 2
area_rect = a * b
area_circle = 3.14 * (r ** 2)
remaining_area = area_rect - area_circle

print(f"{remaining_area:.2f}")
```

```python
# Bài 5
c = input("Nhập chữ cái hoa: ")
# Chuyển sang chữ thường
lower_c = c.lower()

if lower_c == 'a':
    # Trường hợp đặc biệt: trước 'a' coi như là 'z' hoặc xử lý riêng
    print('z') 
else:
    # Lấy ký tự liền trước trong bảng mã ASCII
    print(chr(ord(lower_c) - 1))
```

```python
# Bài 6
import math

a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"{s:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

```python
# Bài 8
ten = input("Ten chu ho: ")
cu = int(input("Chi so thang truoc: "))
moi = int(input("Chi so thang nay: "))

kwh = moi - cu
tien = 0

if kwh <= 50:
    tien = kwh * 1984
elif kwh <= 100:
    tien = 50 * 1984 + (kwh - 50) * 2050
elif kwh <= 200:
    tien = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380
elif kwh <= 300:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998
elif kwh <= 400:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350
else:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460

tong_tien = tien * 1.08  # Cộng thêm 8% VAT
print(f"Ho va ten: {ten}")
print(f"Tien phai tra la: {int(tong_tien)}")
```
