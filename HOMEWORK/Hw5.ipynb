```python
#W6A16
n, m = map(int, input().split())
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

for row in matrix:
    print(" ".join(f"{x:>4}" for x in row))
```

```python
#W6A17
n = int(input())
matrix = []
for _ in range(n):
    matrix.append(list(map(int, input().split())))

main_diag = [matrix[i][i] for i in range(n)]
anti_diag = [matrix[i][n - 1 - i] for i in range(n)]

print(main_diag)
print(anti_diag)
```

```python
#W6A18
line = list(map(int, input().split()))
n, m, k = line[0], line[1], line[2]
matrix = []
for _ in range(n):
    matrix.append(list(map(int, input().split())))

col_sum = sum(row[k] for row in matrix)
print(col_sum)
```

```python
#W6A19
def find_max_index(lst):
    if not lst:
        return -1
    return lst.index(max(lst))
```

```python
#W6A20
def find_first_k(lst, k):
    if k in lst:
        return lst.index(k)
    return -1
```

```python
#W6A21
text = input().split()
word_map = {}
for index, word in enumerate(text):
    if word not in word_map:
        word_map[word] = []
    word_map[word].append(index)

result = {k: tuple(v) for k, v in word_map.items()}
print(result)
```

```python
#W6A22
def get_max_position(lst):
    return lst.index(max(lst))
```

```python
#W6A21
def remove_smaller(lst, x):
    return [num for num in lst if num >= x]
```

```python
#W6A23
def move_zeros(lst):
    zeros = [x for x in lst if x == 0]
    others = [x for x in lst if x != 0]
    return zeros + others
```

```python
#W6A25
def find_missing(lst):
    full_set = set(range(10))
    current_set = set(lst)
    return sorted(list(full_set - current_set))
```

```python
def sum_local_maxima(lst):
    total = 0
    for i in range(1, len(lst) - 1):
        if lst[i-1] < lst[i] and lst[i] > lst[i+1]:
            total += lst[i]
    return total
```

```python
def check_monotonic(lst):
    if len(lst) < 2:
        return "Vo danh"
    
    is_inc = all(lst[i] < lst[i+1] for i in range(len(lst)-1))
    is_dec = all(lst[i] > lst[i+1] for i in range(len(lst)-1))

    if is_inc:
        return "Tang thuc su"
    elif is_dec:
        return "Giam thuc su"
    else:
        return "Vo danh"
```
