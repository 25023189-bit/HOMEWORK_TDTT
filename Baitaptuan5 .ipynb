```python
import math

# W5A1
def Max_Of_Two(a, b):
    if a > b:
        return a
    return b

# W5A2
def swap(a, b):
    return b, a

# W5A3
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

# W5A4
def is_perfect(n):
    if n < 1:
        return False
    sum_divisors = 0
    for i in range(1, n):
        if n % i == 0:
            sum_divisors += i
    return sum_divisors == n

# W5A5
def find_first_k(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

# W5A6
def factorial(n):
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# W5A7
def calculator(num1, num2, operat):
    result = 0
    if operat == '+':
        result = num1 + num2
    elif operat == '-':
        result = num1 - num2
    elif operat == '*':
        result = num1 * num2
    elif operat == '/':
        if num2 == 0:
            return None
        result = num1 / num2
    return round(result, 2)

# W5A8
def hamming_distance(x, y):
    return bin(x ^ y).count('1')

# W5A9
def sum_digits(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total

# W5A10
def is_isomorphic(s1, s2):
    if len(s1) != len(s2):
        return False
    map_s1_s2 = {}
    map_s2_s1 = {}
    for c1, c2 in zip(s1, s2):
        if (c1 in map_s1_s2 and map_s1_s2[c1] != c2) or \
           (c2 in map_s2_s1 and map_s2_s1[c2] != c1):
            return False
        map_s1_s2[c1] = c2
        map_s2_s1[c2] = c1
    return True
```
