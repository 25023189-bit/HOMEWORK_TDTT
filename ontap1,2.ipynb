```python
def unique_sorted_student_ids(list1, list2):
    unique_ids = set(list1) | set(list2)
    return sorted(list(unique_ids))

def top3_frequent_names(names):
    count = {}
    for name in names:
        count[name] = count.get(name, 0) + 1
    sorted_names = sorted(count.items(), key=lambda x: (-x[1], x[0]))
    return [x[0] for x in sorted_names[:3]]

def most_frequent_triple(triples):
    count = {}
    for t in triples:
        count[t] = count.get(t, 0) + 1
    max_freq = 0
    if count:
        max_freq = max(count.values())
    for t, freq in count.items():
        if freq == max_freq:
            return t
    return None

def contains_duplicate(nums):
    return len(nums) != len(set(nums))

def even_numbers_in_matrix(matrix):
    return [x for row in matrix for x in row if x % 2 == 0]

def merge_similar_items(items1, items2):
    mapping = {}
    for v, w in items1:
        mapping[v] = mapping.get(v, 0) + w
    for v, w in items2:
        mapping[v] = mapping.get(v, 0) + w
    return sorted([[v, w] for v, w in mapping.items()])

def can_make_odd_sum(arr):
    has_odd = False
    has_even = False
    for x in arr:
        if x % 2 != 0:
            has_odd = True
        else:
            has_even = True
    if not has_odd:
        return False
    if not has_even:
        return len(arr) % 2 != 0
    return True

def balanced_array(n):
    if n % 4 != 0:
        return "NO"
    result = []
    half = n // 2
    for i in range(1, half + 1):
        result.append(i * 2)
    for i in range(1, half):
        result.append(i * 2 - 1)
    result.append(result[-1] + (result[half-1] - result[-2] + 1) + half) 
    
    evens = [i * 2 for i in range(1, half + 1)]
    odds = [i * 2 - 1 for i in range(1, half)]
    last_odd = sum(evens) - sum(odds)
    odds.append(last_odd)
    return evens + odds
```

```python
def max_and_min(nums):
    if not nums:
        raise ValueError
    max_val = nums[0]
    min_val = nums[0]
    for x in nums:
        if x > max_val:
            max_val = x
        if x < min_val:
            min_val = x
    return (max_val, min_val)

def three_smallest(nums):
    if len(nums) < 3:
        raise ValueError
    first = float('inf')
    second = float('inf')
    third = float('inf')
    for x in nums:
        if x < first:
            third = second
            second = first
            first = x
        elif x < second:
            third = second
            second = x
        elif x < third:
            third = x
    return [first, second, third]

def second_largest_and_smallest(nums):
    distinct_nums = list(set(nums))
    if len(distinct_nums) < 2:
        raise ValueError
    distinct_nums.sort()
    return (distinct_nums[-2], distinct_nums[1])

def is_strictly_increasing(nums):
    if len(nums) <= 1:
        return True
    for i in range(len(nums) - 1):
        if nums[i] >= nums[i+1]:
            return False
    return True

def linear_search(nums, target):
    for i in range(len(nums)):
        if nums[i] == target:
            return i
    return -1

def selection_sort(arr):
    nums = arr[:]
    n = len(nums)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if nums[j] < nums[min_idx]:
                min_idx = j
        nums[i], nums[min_idx] = nums[min_idx], nums[i]
    return nums

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)
```
