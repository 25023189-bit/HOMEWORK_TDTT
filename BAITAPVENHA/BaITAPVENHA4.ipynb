```python
# W4A6
import math

def check_prime(num):
    if num < 2:
        return False
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True

try:
    a, b = map(int, input().split())
    total = 0
    for x in range(a, b + 1):
        if check_prime(x):
            total += x
    print(total)
except:
    pass
```

```python
# W4A8
def get_reverse(num):
    rev = 0
    temp = num
    while temp > 0:
        rev = rev * 10 + temp % 10
        temp //= 10
    return rev

n = int(input())
steps = 0
while True:
    rev_n = get_reverse(n)
    if n == rev_n:
        break
    n = n + rev_n
    steps += 1

print(steps, n)
```

```python
# W4A9
n = int(input())
found = False
i = 0
while True:
    sq = i * i
    if sq > n:
        break
    
    s_sq = str(sq)

    if len(set(s_sq)) == len(s_sq):
        print(sq, end=' ')
        found = True
    i += 1

if not found:
    print("(no number)")
else:
    print()
```

```python
# W4A10
def collatz_len(x):
    length = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3 * x + 1
        length += 1
    return length

n = int(input())
max_len = 0
best_x = 1

for x in range(1, n + 1):
    curr_len = collatz_len(x)
    if curr_len > max_len:
        max_len = curr_len
        best_x = x
    elif curr_len == max_len:
     
        pass

print(best_x, max_len)
```

```python
# W4A11
import math

n = int(input())
count = 0
for i in range(1, int(math.sqrt(n)) + 1):
    if n % i == 0:
        if i % 2 == 0:
            count += 1
        
        div2 = n // i
        if div2 != i and div2 % 2 == 0:
            count += 1
print(count)
```

```python
# W4A13
import math

def sum_proper_divisors(num):
    if num < 2:
        return 0
    s = 1
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            s += i
            if i * i != num:
                s += num // i
    return s

try:
    line = input().split()
    if len(line) == 2:
        a, b = map(int, line)
        if sum_proper_divisors(a) == b and sum_proper_divisors(b) == a:
            print("true")
        else:
            print("false")
except:
    print("false")
```
