```python
#W9A1
def merge_lists(list1, list2):
    res = []
    i, j = 0, 0
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            res.append(list1[i])
            i += 1
        else:
            res.append(list2[j])
            j += 1
    res.extend(list1[i:])
    res.extend(list2[j:])
    return res

try:
    s1 = input().split()
    s2 = input().split()
    list1 = [int(x) for x in s1]
    list2 = [int(x) for x in s2]
    print(*(merge_lists(list1, list2)))
except:
    pass
```

```python
#W9A2
import heapq

def kth_smallest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])
    min_heap = []
    for r in range(min(n, k)):
        heapq.heappush(min_heap, (matrix[r][0], r, 0))
    
    ans = -1
    for _ in range(k):
        ans, r, c = heapq.heappop(min_heap)
        if c + 1 < m:
            heapq.heappush(min_heap, (matrix[r][c+1], r, c+1))
    return ans

try:
    lines = []
    while True:
        try:
            line = input()
            if not line: break
            lines.append(line)
        except EOFError:
            break
            
    if lines:
        k = int(lines[-1])
        matrix = []
        for line in lines[:-1]:
            row = list(map(int, line.split()))
            if row:
                matrix.append(row)
        total_elements = sum(len(row) for row in matrix)
        target = total_elements - k + 1
        print(kth_smallest(matrix, target))
except:
    pass
```

```python
#W9A3
def minimize_number(n_str):
    s = list(n_str)
    best = int(n_str)
    for i in range(len(s)):
        for j in range(i + 1, len(s)):
            s[i], s[j] = s[j], s[i]
            if s[0] != '0': 
                val = int("".join(s))
                if val < best:
                    best = val
            s[i], s[j] = s[j], s[i]
    return best

try:
    while True:
        try:
            line = input().strip()
            if not line: break
            print(minimize_number(line))
        except EOFError:
            break
except:
    pass
```

```python
#W9A4
def check_transform(a, b):
    i, j = 0, 0
    while i < len(a) and j < len(b):
        if a[i] == b[j]:
            i += 1
            j += 1
        elif a[i].upper() == b[j]:
            i += 1
            j += 1
        elif a[i].islower():
            i += 1
        else:
            return "NO"
    
    while i < len(a):
        if a[i].isupper():
            return "NO"
        i += 1
        
    return "YES" if j == len(b) else "NO"

try:
    Q = int(input())
    for _ in range(Q):
        a = input().strip()
        b = input().strip()
        print(check_transform(a, b))
except:
    pass
```

```python
#W9A5
def is_sublist(list1, list2):
    n, m = len(list1), len(list2)
    if n == 0: return "YES"
    if m < n: return "NO"
    
    for i in range(m - n + 1):
        if list2[i : i + n] == list1:
            return "YES"
    return "NO"

try:
    l1 = list(map(int, input().split()))
    l2 = list(map(int, input().split()))
    print(is_sublist(l1, l2))
except:
    pass
```

```python
#W9A6
def next_permutation(nums):
    i = len(nums) - 2
    while i >= 0 and nums[i] >= nums[i + 1]:
        i -= 1
    
    if i >= 0:
        j = len(nums) - 1
        while nums[j] <= nums[i]:
            j -= 1
        nums[i], nums[j] = nums[j], nums[i]
    
    left, right = i + 1, len(nums) - 1
    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1
    return nums

try:
    n = int(input())
    arr = list(map(int, input().split()))
    res = next_permutation(arr)
    print(*(res))
except:
    pass
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    count = 0
    arr = [0] + flowerbed + [0]
    for i in range(1, len(arr) - 1):
        if arr[i-1] == 0 and arr[i] == 0 and arr[i+1] == 0:
            arr[i] = 1
            count += 1
    return count >= k
```

```python
#W9A8
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}
    for move in moves:
        if move == 'U': y += 1
        elif move == 'D': y -= 1
        elif move == 'R': x += 1
        elif move == 'L': x -= 1
        
        if (x, y) in visited:
            return True
        visited.add((x, y))
    return False
```

```python
#W9A9
def Median(nums):
    a, b, c, d, e = nums[0], nums[1], nums[2], nums[3], nums[4]
    
    if a > b: a, b = b, a
    if c > d: c, d = d, c
    if a > c: 
        a, c = c, a
        b, d = d, b
        
    if b > e: b, e = e, b
    
    if b < c:
        if c < e: return c
        else: return e
    else:
        if b < d: return b
        else: return d
```

```python
#W9A10
import bisect

def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    radius = 0
    
    for house in houses:
        idx = bisect.bisect_left(heaters, house)
        dist1 = float('inf')
        dist2 = float('inf')
        
        if idx < len(heaters):
            dist1 = heaters[idx] - house
        if idx > 0:
            dist2 = house - heaters[idx - 1]
            
        radius = max(radius, min(dist1, dist2))
        
    return radius
```
