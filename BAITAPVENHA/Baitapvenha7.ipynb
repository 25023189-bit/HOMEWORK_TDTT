```python
#W7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

try:
    arr = list(map(int, input().split()))
    print(*bubble_sort(arr))
except:
    pass
```

```python
#W7A5
def count_pairs(arr, X):
    count = 0
    seen = {}
    for num in arr:
        target = X - num
        if target in seen:
            count += seen[target]
        seen[num] = seen.get(num, 0) + 1
    return count

try:
    arr = list(map(int, input().split()))
    X = int(input())
    print(count_pairs(arr, X))
except:
    pass
```

```python
#W7A6
import ast

def flatten(nested_list):
    flat_list = []
    for item in nested_list:
        if isinstance(item, list):
            flat_list.extend(flatten(item))
        else:
            flat_list.append(item)
    return flat_list

try:
    inp = input()
    nested_list = ast.literal_eval(inp)
    print(flatten(nested_list))
except:
    pass
```

```python
#W7A7
import bisect

def length_of_lis(nums):
    if not nums:
        return 0
    sub = []
    for x in nums:
        idx = bisect.bisect_left(sub, x)
        if idx == len(sub):
            sub.append(x)
        else:
            sub[idx] = x
    return len(sub)

try:
    nums = list(map(int, input().split()))
    print(length_of_lis(nums))
except:
    pass
```

```python
#W7A8
import ast

def solve_intervals(intervals, queries):
    results = []
    for q in queries:
        min_len = float('inf')
        found = False
        for interval in intervals:
            l, r = interval
            if l <= q <= r:
                found = True
                curr_len = r - l + 1
                if curr_len < min_len:
                    min_len = curr_len
        if found:
            results.append(min_len)
        else:
            results.append(-1)
    return results

try:
    intervals = ast.literal_eval(input())
    queries = ast.literal_eval(input())
    print(solve_intervals(intervals, queries))
except:
    pass
```

```python
#W7A9
import bisect

def count_smaller(nums):
    counts = []
    seen = []
    for x in reversed(nums):
        idx = bisect.bisect_left(seen, x)
        counts.append(idx)
        bisect.insort(seen, x)
    return counts[::-1]

try:
    nums = list(map(int, input().split()))
    print(count_smaller(nums))
except:
    pass
```

```python
#W7A10
import bisect

def max_subarray_sum_mod(nums, m):
    max_mod = 0
    prefix = 0
    sorted_prefixes = [0]
    
    for x in nums:
        prefix = (prefix + x) % m
        max_mod = max(max_mod, prefix)
        
        idx = bisect.bisect_right(sorted_prefixes, prefix)
        if idx < len(sorted_prefixes):
            max_mod = max(max_mod, (prefix - sorted_prefixes[idx] + m) % m)
            
        bisect.insort(sorted_prefixes, prefix)
        
    return max_mod

try:
    nums = list(map(int, input().split()))
    m = int(input())
    print(max_subarray_sum_mod(nums, m))
except:
    pass
```

```python
#W7A13
from collections import Counter

def longest_harmonious_subsequence(nums):
    count = Counter(nums)
    max_len = 0
    for x in count:
        if x + 1 in count:
            max_len = max(max_len, count[x] + count[x+1])
    return max_len

try:
    nums = list(map(int, input().split()))
    print(longest_harmonious_subsequence(nums))
except:
    pass
```
