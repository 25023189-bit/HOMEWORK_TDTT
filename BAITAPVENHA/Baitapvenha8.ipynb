```python
#W7A16
from collections import Counter

try:
    arr = list(map(int, input().split()))
    counts = Counter(arr)
    remove_count = 0
    
    for x, freq in counts.items():
        if freq > x:
            remove_count += freq - x
        elif freq < x:
            remove_count += freq
            
    print(remove_count)
except:
    pass
```

```python
#W7A17
def find_majority(arr):
    candidate = None
    count = 0
    for num in arr:
        if count == 0:
            candidate = num
            count = 1
        elif num == candidate:
            count += 1
        else:
            count -= 1
            
    if arr.count(candidate) > len(arr) // 2:
        return candidate
    return None

try:
    arr = list(map(int, input().split()))
    res = find_majority(arr)
    if res is not None:
        print(res)
except:
    pass
```

```python
#W7A19
try:
    arr1 = list(map(int, input().split()))
    arr2 = list(map(int, input().split()))
    
    res = []
    i, j = 0, 0
    
    # Kiem tra chieu tang hay giam
    ascending = True
    if (len(arr1) > 1 and arr1[0] > arr1[-1]) or (len(arr2) > 1 and arr2[0] > arr2[-1]):
        ascending = False
        
    while i < len(arr1) and j < len(arr2):
        if ascending:
            if arr1[i] < arr2[j]:
                res.append(arr1[i])
                i += 1
            else:
                res.append(arr2[j])
                j += 1
        else:
            if arr1[i] > arr2[j]:
                res.append(arr1[i])
                i += 1
            else:
                res.append(arr2[j])
                j += 1
                
    res.extend(arr1[i:])
    res.extend(arr2[j:])
    print(*res)
except:
    pass
```

```python
#W7A25
try:
    n_str = input().strip()
    k = int(input())
    
    stack = []
    to_remove = len(n_str) - k
    
    for digit in n_str:
        while to_remove > 0 and stack and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
        stack.append(digit)
        
    print("".join(stack[:k]))
except:
    pass
```

```python
#W7A27
def is_valid_parentheses(s):
    stack = []
    mapping = {")": "(", "}": "{", "]": "["}
    for char in s:
        if char in mapping:
            top_element = stack.pop() if stack else '#'
            if mapping[char] != top_element:
                return "NO"
        elif char in mapping.values():
            stack.append(char)
    return "NO" if stack else "YES"

try:
    s = input()
    print(is_valid_parentheses(s))
except:
    pass
```

```python
#W7A28
def is_composite(n):
    if n <= 3:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return True
    return False

try:
    N = int(input())
    max_len = 0
    best_seq = []
    
    current_seq = []
    for i in range(1, N): # Theo logic strict less than N tu output mau
        if is_composite(i):
            current_seq.append(i)
        else:
            if len(current_seq) > max_len:
                max_len = len(current_seq)
                best_seq = current_seq[:]
            current_seq = []
            
    if len(current_seq) > max_len:
        best_seq = current_seq
        
    if not best_seq:
        print()
    else:
        print(", ".join(map(str, best_seq)))
except:
    pass
```

```python
#W7A30
import bisect

try:
    arr = list(map(int, input().split()))
    if not arr:
        print(0)
    else:
        sub = []
        for x in arr:
            i = bisect.bisect_left(sub, x)
            if i == len(sub):
                sub.append(x)
            else:
                sub[i] = x
        print(len(sub))
except:
    pass
```
