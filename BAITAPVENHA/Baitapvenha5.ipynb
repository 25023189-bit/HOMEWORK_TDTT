```python
# W5A1
def max_of_two(a, b):
    if a > b:
        return a
    return b
```

```python
# W5A2
def swap(a, b):
    return b, a
```

```python
# W5A3
import math

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

```python
# W5A4
def is_perfect(n):
    if n < 2:
        return False
    total = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            total += i
            if i*i != n:
                total += n // i
    return total == n
```

```python
# W5A5
def find_first_occurrence(lst, k):
    try:
        return lst.index(k)
    except ValueError:
        return -1
```

```python
# W5A6
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
```

```python
# W5A7
def calculator(num1, num2, operator):
    if operator == '+':
        return round(num1 + num2, 2)
    elif operator == '-':
        return round(num1 - num2, 2)
    elif operator == '*':
        return round(num1 * num2, 2)
    elif operator == '/':
        if num2 == 0:
            return "Error"
        return round(num1 / num2, 2)
    else:
        return "Invalid operator"
```

```python
# W5A8
def hamming_distance(x, y):
    val = x ^ y
    dist = 0
    while val > 0:
        if val % 2 == 1:
            dist += 1
        val //= 2
    return dist
```

```python
# W5A9
def sum_digits(n):
    n = abs(n)
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
```

```python
# W5A10
def is_isomorphic(s1, s2):
    if len(s1) != len(s2):
        return False
    
    map_s1_s2 = {}
    map_s2_s1 = {}
    
    for c1, c2 in zip(s1, s2):
        if c1 in map_s1_s2:
            if map_s1_s2[c1] != c2:
                return False
        else:
            map_s1_s2[c1] = c2
            
        if c2 in map_s2_s1:
            if map_s2_s1[c2] != c1:
                return False
        else:
            map_s2_s1[c2] = c1
            
    return True
```
