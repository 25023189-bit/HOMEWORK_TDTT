```python
#W6A4
items = input().split()
result = {}
for item in items:
    key, value = item.split(':')
    if key not in result:
        result[key] = []
    result[key].append(value)
print(result)
```

```python
#W6A5
lst = list(map(int, input().split()))
pos = 0
neg = 0
zeros = 0
for x in lst:
    if x > 0:
        pos += 1
    elif x < 0:
        neg += 1
    else:
        zeros += 1
print({'positives': pos, 'negatives': neg, 'zeros': zeros})
```

```python
#W6A9
def input_dict():
    items = input().split()
    d = {}
    for item in items:
        k, v = item.split(':')
        d[k] = int(v)
    return d

d1 = input_dict()
d2 = input_dict()
result = d1.copy()

for k, v in d2.items():
    if k in result:
        result[k] += v
    else:
        result[k] = v

sorted_result = dict(sorted(result.items()))
print(sorted_result)
```

```python
#W6A10
lst = list(map(int, input().split()))
k = int(input())
pairs = []
seen = set()
for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
        if lst[i] + lst[j] == k:
            pair = tuple(sorted((lst[i], lst[j])))
            if pair not in seen:
                pairs.append(pair)
                seen.add(pair)
pairs.sort(key=lambda x: x[0])
print(pairs)
```

```python
#W6A12
lst = list(map(int, input().split()))
counts = {}
for x in lst:
    counts[x] = counts.get(x, 0) + 1

max_freq = -1
if counts:
    max_freq = max(counts.values())

candidates = [k for k, v in counts.items() if v == max_freq]
if candidates:
    print(min(candidates))
```
