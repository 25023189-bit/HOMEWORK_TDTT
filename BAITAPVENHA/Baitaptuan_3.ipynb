```python
# W3A11
a = int(input())
b = int(input())
c = int(input())

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giác đều")
    elif a == b or b == c or a == c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
else:
    print("Không phải tam giác")
```

```python
# W3A16
x = float(input())

floor_val = int(x) if x >= 0 or x == int(x) else int(x) - 1
ceil_val = int(x) if x <= 0 or x == int(x) else int(x) + 1
round_val = int(x + 0.5) if x >= 0 else int(x - 0.5)

print(ceil_val, floor_val, round_val)
```

```python
# W3A17
line = list(map(int, input().split()))
if len(line) == 4:
    a, b, c, d = line
    if a != 0 and b % a == 0:
        q = b // a
        if b * q == c and c * q == d:
            print("Yes")
        else:
            print("No")
    else:
        print("No")
```
