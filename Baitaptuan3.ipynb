```python
#W3A3
def is_power_of_two(n):
    return n > 0 and (n & (n - 1)) == 0
nums = [1, 2, 3, 4, 8, 10, 16, 18]

for n in nums:
    if is_power_of_two(n):
        print(f"{n} là lũy thừa của 2")
    else:
        print(f"{n} không phải lũy thừa của 2")
```

```python
n = float(input())

# Làm tròn xuống
if n >= 0:
    floor = n - (n % 1)
else:
    floor = n - (n % 1) - (1 if n % 1 != 0 else 0)

# Làm tròn lên
if n >= 0:
    ceil = n - (n % 1) + (1 if n % 1 != 0 else 0)
else:
    ceil = n - (n % 1)


if n >= 0:
    round_nearest = n - (n % 1) + (1 if n % 1 >= 0.5 else 0)
else:
    round_nearest = n - (n % 1) - (1 if (-n % 1) >= 0.5 else 0)

print(floor, ceil, round_nearest)
```
