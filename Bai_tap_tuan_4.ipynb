```python
# W4A2
import math
n = -1
while True:
    try:
        inp = int(input())
        if inp > 0:
            n = inp
            break
    except ValueError:
        continue

if n < 2:
    print("False") 
else:
    is_prime = True
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            is_prime = False
            break
    if is_prime:
        print("True")
    else:
        print("False")
```

```python
# W4A6
def is_prime(num):
    if num < 2: return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0: return False
    return True

try:
    line = input().split()
    a, b = int(line[0]), int(line[1])
    total = 0
    for i in range(a, b + 1):
        if is_prime(i):
            total += i
    print(total)
except ValueError:
    pass
```

```python
# W4A7
try:
    n = int(input())
    if n >= 2:
        max_prime = -1
        while n % 2 == 0:
            max_prime = 2
            n //= 2
        f = 3
        while f * f <= n:
            while n % f == 0:
                max_prime = f
                n //= f
            f += 2
        if n > 2:
            max_prime = n
        print(max_prime)
except ValueError:
    pass
```

```python
# W4A8
def rev(num):
    r = 0
    while num > 0:
        r = r * 10 + num % 10
        num //= 10
    return r

try:
    n = int(input())
    steps = 0
    while n != rev(n):
        n = n + rev(n)
        steps += 1
    print(f"{steps} {n}")
except ValueError:
    pass
```

```python
# W4A9
try:
    n = int(input())
    i = 0
    found = False
    res = []
    while True:
        sq = i * i
        if sq > n: break
        s = str(sq)
        if len(set(s)) == len(s):
            res.append(s)
            found = True
        i += 1
    
    if not found:
        print("(no number)")
    else:
        print(" ".join(res))
except ValueError:
    pass
```

```python
# W4A10
def collatz_len(x):
    l = 1
    while x != 1:
        if x % 2 == 0: x //= 2
        else: x = 3 * x + 1
        l += 1
    return l

try:
    n = int(input())
    max_l = 0
    res_x = 1
    for x in range(1, n + 1):
        cl = collatz_len(x)
        if cl > max_l:
            max_l = cl
            res_x = x
    print(f"{res_x} {max_l}")
except ValueError:
    pass
```

```python
# W4A13
def sum_div(num):
    s = 0
    for i in range(1, int(num**0.5) + 1):
        if num % i == 0:
            s += i
            if i != 1 and i * i != num:
                s += (num // i)
    return s

try:
    line = input().split()
    n1, n2 = int(line[0]), int(line[1])
    if sum_div(n1) == n2 and sum_div(n2) == n1:
        print("true")
    else:
        print("false")
except ValueError:
    pass
```
